```python
pitcher_team_cleaned = []
for element in hurlers:
    
    element = element.replace('\n',' ')
    
    length = len(element.split())
    
    if length < 4 or length > 5:
    
        print('Format not what expected')
    
    n = length - 2
    
    groups = element.split(' ',n)
    
    team  = ' '.join(groups[:n])

    pitcher = ' '.join(groups[n:])

    pitcher_team_cleaned.append((team,pitcher))

pitcher_team_cleaned
```

```python
df = pd.DataFrame(data = pitcher_team_cleaned, columns=['Team', 'Pitcher'])
df
```
