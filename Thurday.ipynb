```python
# Build a dictionary of all words
def build_dict():
    with open('words.txt') as my_file:
    #wordlist = data.split() # split the data to form a list and save it in the wordlist
        words = {}
        for word in my_file:
            word = word.strip()
            if (word in words):
                words[word] = words[word] + 1
            else:
                words[word] = 1
   
    return words
```

```python
def isreversewordindict(word,reverseword, words):    
    if (reverseword in words):
        return [word, reverseword]
    return None
```

```python
def find_all_words(words):
    res = []
    reversewordlist = []
    for w in words:
        reverseword  = w[::-1]
        if w not in reversewordlist:
            newwordtoadd = isreversewordindict(w,reverseword, words)
            # newwordtoadd will be None or a word
            if (newwordtoadd ):
                res.append(newwordtoadd)
                reversewordlist.append(reverseword)
    return res
```

```python
%%time
mylist = find_all_words(build_dict())
print (len(mylist))
print (mylist[12])
```
